```python
file=open("/content/drive/MyDrive/audio_path.txt","r")
text=open("/content/drive/MyDrive/transcripts_path.txt","r")
error=0
count=0

end=True
while (end==True):
     src=file.readline()
     txt_path=text.readline()
     print("TRANSCRIPT PATH: ",txt_path)
     '''if "\n" in txt_path:
         txt_path=txt_path.replace("\n","")
     txt_file=open(txt_path,"r")
     txt=txt_file.read()
     print(txt)'''
     if "\n" in txt_path:
         txt_path=txt_path.replace("\n","")
     if "\ufeff" in txt_path:
          txt_path=txt_path.replace("\ufeff","")
     txt_file=open(txt_path,"r")
     txt=txt_file.read()
     print("THE TRANSCRIPT IS ",txt)
     print(src)
     if "\n" in src:
         src=src.replace("\n","")
     else:
         end=False
     print("AUDIO PATH IS: ",src)
     speech, rate = librosa.load(src,sr=16000, mono= True)
     input_values = tokenizer(speech, return_tensors = 'pt').input_values
     logits = model(input_values).logits
     predicted_ids = torch.argmax(logits, dim =-1)
     transcriptions = tokenizer.decode(predicted_ids[0])
     print("TRANSCRIBED TRANSCRIPTIONS: ",transcriptions)
     wer=fastwer.score_sent(txt,transcriptions,char_level=True)
     error+=wer
     count+=1
     print("EORD ERROR RATE: ",wer)
print(wer/count)
```
